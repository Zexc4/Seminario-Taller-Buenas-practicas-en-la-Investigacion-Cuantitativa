## 1.2. Análisis Cuantitativo de Riesgos: Matriz de Probabilidad de Impacto

La Matriz de Probabilidad de Impacto es una técnica clásica de evaluación de riesgos utilizada para representar de forma cuantitativa eventos inciertos que puedan afectar un proyecto experimental, una simulación o un sistema.

Esta matriz se construye combinando dos dimensiones:

* Probabilidad de ocurrencia del evento (Baja, Media, Alta)
* Impacto si el evento ocurre (Leve, Moderado, Crítico)

Cada combinación genera un valor numérico, permitiendo calcular un nivel de riesgo para priorizar acciones.

<a id="1.2.1"></a>
### 1.2.1. Matriz 3x3: Estructura Típica

| Probabilidad / Impacto | Leve (1) | Moderado (2) | Crítico (3) |
| ---------------------- | -------- | ------------ | ----------- |
| **Baja (1)**           | 1        | 2            | 3           |
| **Media (2)**          | 2        | 4            | 6           |
| **Alta (3)**           | 3        | 6            | 9           |


> Valores más altos indican mayor prioridad de mitigación.

<a id="1.2.2"></a>
### 1.2.2. Implementación en Python

> La plataforma [PreventionWeb](https://www.preventionweb.net/understanding-disaster-risk/disaster-losses-and-statistics/global-risk-data-sets), vinculada a iniciativas como INFORM y GIRI, ofrece datos sobre riesgos globales en infraestructura, desastres naturales, inundaciones, terremotos, entre otros. Estos datos están en formato abierto y son ampliamente utilizados en estudios científicos sobre gestión de riesgos y resiliencia

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulación de dataset basado en datos reales (gestión de riesgos)
df = pd.DataFrame({
    'Evento': ['Terremoto Menor', 'Terremoto Moderado', 'Terremoto Mayor'],
    'Probabilidad': ['Alta', 'Media', 'Baja'],
    'Impacto': ['Crítico', 'Moderado', 'Leve']
})

# Mapear escalas cualitativas a cuantitativas
mapping = {'Baja':1, 'Media':2, 'Alta':3, 'Leve':1, 'Moderado':2, 'Crítico':3}
df['P_num'] = df['Probabilidad'].map(mapping)
df['I_num'] = df['Impacto'].map(mapping)

# Cálculo del riesgo
df['Riesgo'] = df['P_num'] * df['I_num']

# Pivot table
matrix = df.pivot(index='Probabilidad', columns='Impacto', values='Riesgo')
matrix = matrix.reindex(index=['Baja','Media','Alta'], columns=['Leve','Moderado','Crítico'])

# Visualización con formato flotante redondeado
plt.figure(figsize=(6, 4))
sns.heatmap(matrix, annot=True, cmap='Reds', fmt='.0f')  # <-- cambio aquí
plt.title('Matriz de Probabilidad vs Impacto (Datos reales)')
plt.xlabel('Impacto')
plt.ylabel('Probabilidad')
plt.tight_layout()
plt.show()
````

<a id="1.2.3"></a>
### 1.2.3. Interpretación Cuantitativa


* Esta matriz convierte juicios cualitativos en valores numéricos objetivos, lo que permite tomar decisiones cuantitativamente justificadas.
* Es aplicable para priorizar variables, analizar riesgos experimentales o definir escenarios críticos en simulaciones.
* Su uso es común en ingeniería de proyectos, modelado de sistemas y validación de experimentos.
* Usa datos numéricos y reglas consistentes.
* Permite reproducibilidad y comparación de resultados.
* Es una herramienta previa a modelos estadísticos o de optimización.
* Puentea lo cualitativo y lo cuantitativo, representando lo subjetivo con valores escalables.

> _Después de cuantificar riesgos, otra tarea inicial en un análisis cuantitativo es organizar y representar datos medidos. Para esto, se utilizan reglas como Sturges, que ayudan a construir histogramas óptimos y detectar patrones._

<a id="1.3"></a>
## 1.3. Clasificación de Datos: Aplicación de la Regla de Sturges

Cuando se recolectan datos experimentales o simulados (por ejemplo, expectativas de observables cuánticos, tiempos de respuesta en un sistema, tasas de crecimiento), una de las primeras tareas del análisis cuantitativo es:

* Agrupar los datos en clases o intervalos.
* Construir histogramas para explorar la distribución.
* Detectar asimetrías, sesgos, outliers o patrones ocultos.

> Pero… ¿cómo saber cuántas clases usar en un histograma?

_Ahí es donde entra la Regla de Sturges._


<a id="1.3.1"></a>
### 1.3.1. Regla de Sturges: Definición

La Regla de Sturges es una fórmula empírica que estima el número óptimo de clases (bins) para representar un conjunto de datos en un histograma.

<div align="center">
$k=1+log_2(n)$
</div>

Donde:
* $k:$  número recomendado de clases
* $n:$  número de observaciones

También puede expresarse como:

<div align="center">
$k=[log_2(n)+1]$
</div>

Esto permite evitar subrepresentar o sobrecargar los histogramas, y obtener una visión clara de la distribución.

<a id="1.3.2"></a>
### 1.3.2. Implementación en Python

> Usaremos un dataset real: el famoso Iris dataset, que contiene medidas continuas como longitud del sépalo, longitud del pétalo, etc. Puedes descargarlo desde [UCI Machine Learning Repository](https://www.kaggle.com/datasets/uciml/iris) o usar directamente la URL público.


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
import pandas as pd

# Cargar Iris dataset desde GitHub
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# Variable continua: sepal_length
datos = df['sepal_length']

n = len(datos)
k = math.ceil(math.log2(n) + 1)
print(f"Número de clases según Sturges: {k}")

plt.figure(figsize=(8,5))
sns.histplot(datos, bins=k, kde=True, color='teal')
plt.title(f'Histograma de Longitud de Sépalo (k = {k} clases)')
plt.xlabel('Longitud de Sépalo (cm)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()
````

<a id="1.3.3"></a>
### 1.3.3. Interpretación Cuantitativa

* Esta técnica es objetiva, reproducible y basada en el tamaño muestral, lo cual es clave para estudios experimentales o simulaciones.
* Evita la subjetividad en la selección de clases, que podría alterar la interpretación visual de los datos.
* Es especialmente útil en fases exploratorias de análisis cuantitativo, como previo al PCA, regresiones, clustering, o ANOVA.
* Sistematiza la visualización inicial de datos.
* Permite hacer inferencias preliminares sobre la forma de la distribución.
* Sienta las bases para análisis más complejos (test de normalidad, selección de variables, validación de supuestos).

<a id="1.3.4"></a>
### 1.3.4. Comparación con otros métodos

Otras reglas para determinar el número de clases incluyen:

| Método                | Fórmula                              | Comentario                          |
| --------------------- | ------------------------------------ | ----------------------------------- |
| **Sturges**           | $k = 1 + \log_2(n)$                  | Estándar, ideal para datos normales |
| **Freedman–Diaconis** | $h = 2 \cdot IQR / \sqrt[3]{n}$      | Más robusta ante outliers           |
| **Scott**             | $h = 3.5 \cdot \sigma / \sqrt[3]{n}$ | Basada en desviación estándar       |

## 1.4. Diferencias entre el Análisis Cuantitativo y Cualitativo

En el diseño de proyectos científicos y de ingeniería, el enfoque de análisis elegido —cuantitativo o cualitativo— determina el tipo de evidencia que se obtiene, la forma de estructurar los datos y el método de interpretación de los resultados.



A continuación, se contrastan ambos enfoques usando dos herramientas como referencia práctica:

* La Matriz de Impacto, que nace en un contexto más cualitativo pero puede formalizarse cuantitativamente.
* La Regla de Sturges, que es una técnica completamente cuantitativa.

<a id="1.4.1"></a>
### 1.4.1. Comparación Conceptual


| Característica              | Análisis Cuantitativo                       | Análisis Cualitativo                          |
| --------------------------- | ------------------------------------------- | --------------------------------------------- |
| **Naturaleza de los datos** | Numéricos, medibles, objetivos              | Categóricos, subjetivos, descriptivos         |
| **Tipo de razonamiento**    | Deductivo, hipotético-deductivo             | Inductivo o interpretativo                    |
| **Enfoque**                 | Explicación de fenómenos mediante variables | Comprensión profunda del fenómeno             |
| **Instrumentos comunes**    | Encuestas, sensores, simulaciones, modelos  | Entrevistas, observaciones, análisis textual  |
| **Resultados**              | Estadísticos, generalizables, replicables   | Ricos en contexto, no generalizables          |
| **Ejemplo práctico**        | Regla de Sturges (análisis de distribución) | Juicio experto al construir matriz de impacto |


<a id="1.4.2"></a>
### 1.4.2. Aplicación a la Matriz de Probabilidad de Impacto

| Etapa de uso              | Cualitativo                                 | Cuantitativo                                       |
| ------------------------- | ------------------------------------------- | -------------------------------------------------- |
| **Construcción inicial**  | Requiere juicio experto para estimar riesgo | Puede usarse historial de datos reales             |
| **Asignación de escalas** | Baja, media, alta (descripción subjetiva)   | Valores 1, 2, 3 (escala ordinal cuantificada)      |
| **Propósito**             | Visualizar riesgo percibido                 | Priorizar acciones con base en criterios numéricos |
| **Ejemplo de transición** | Estimar que un riesgo es "alto impacto"     | Asignar valor 3 de impacto → Riesgo = Prob × 3     |


> La matriz puede comenzar como una herramienta cualitativa, pero al cuantificar sus elementos, se convierte en una técnica de análisis cuantitativo reproducible.

<a id="1.4.3"></a>
### 1.4.3. Aplicación a la Regla de Sturges

| Etapa de uso                | Cualitativo                                  | Cuantitativo                                       |
| --------------------------- | -------------------------------------------- | -------------------------------------------------- |
| **Motivación**              | Deseo de visualizar la distribución de datos | Representación estructurada y optimizada           |
| **Número de clases (bins)** | Elegido arbitrariamente                      | Calculado con fórmula basada en el tamaño muestral |
| **Análisis visual**         | Gráfico subjetivo o no representativo        | Histograma balanceado y comparable                 |
| **Ejemplo**                 | “Se ve bien con 5 clases”                    | “Según Sturges, se requieren 7 clases”             |


> El análisis cualitativo puede guiarnos a explorar un fenómeno, pero la formalización cuantitativa permite objetividad, comparación y reproducibilidad, características esenciales del análisis científico.

<a id="1.4.3"></a>
### 1.4.4. Conexión entre ambos enfoques

Aunque diferentes, los enfoques cualitativo y cuantitativo no son excluyentes, sino complementarios.

Por ejemplo:

* En la matriz de impacto, el juicio experto (cualitativo) inicia el análisis, pero es la cuantificación la que permite su uso en simulaciones y modelos de decisión.
* En la clasificación de datos, la intuición puede guiar el análisis, pero son las fórmulas como la de Sturges las que dan precisión y rigor.

<a id="1.4.5"></a>
### 1.4.5. Conclusión

Las herramientas vistas —la Matriz de Impacto y la Regla de Sturges— ilustran la evolución natural del pensamiento científico:

1. Se parte de observaciones y percepciones cualitativas.
2. Se estructuran en modelos cuantitativos.
3. Se validan, visualizan y comunican los resultados mediante métricas objetivas.

> Este tránsito es el fundamento del análisis cuantitativo aplicado, especialmente en disciplinas como la ingeniería, la física computacional, la gestión de riesgos o la ciencia de datos.

## 2.1. Investigación Experimental

Es el tipo más riguroso de investigación cuantitativa. Se caracteriza por:

* Manipulación directa de la VI.
* Control estricto de las variables externas.
* Asignación aleatoria de sujetos a grupos.

🎯 Su objetivo es establecer relaciones de **causalidad**.

> Evaluar el efecto de un conservante natural (como extracto vegetal) en la inhibición del crecimiento bacteriano, midiendo la absorbancia (VD) a lo largo de los días.

## 2.2. Investigación Cuasi-experimental

Se utiliza cuando no es posible aplicar aleatoriedad completa. Por ejemplo:

* Grupos ya formados (escuelas, plantas industriales, comunidades).
* Menor control de variables externas, pero aún se manipula la VI.


👨‍🏫 **Ejemplo:**

> Comparar el rendimiento de productos procesados en distintas líneas de producción, sin posibilidad de asignar condiciones aleatoriamente.

## 2.3. Investigación Cuantitativa No Experimental

No hay manipulación de variables. Se basa en la observación, registro y análisis estadístico de fenómenos existentes.

* **_Descriptiva:_** Describe características de una población o fenómeno.
* **_Correlacional:_** Examina relaciones entre variables sin establecer causalidad.

💧 **Ejemplo:**

> Analizar la relación entre turbidez, conductividad y pH en cuerpos de agua de distintas regiones.

🧠 Aquí, **PCA** es útil para reducir variables y encontrar las más relevantes o representativas del comportamiento global.

<div align="center">


🧩 **iNTERPRETACIÓN:**

* PC1 y PC2 son combinaciones de las variables originales (turbidez, conductividad, pH).
* Los grupos separados por región indican que ciertas regiones tienen características fisicoquímicas distintas.
* Puedes identificar qué variables contribuyen más a cada componente analizando pca.components_.

## 3.3. Ejemplo 3: Aplicación en Ciencia e Ingeniería

* Uso de estas técnicas para validar sensores, materiales, procesos industriales.

Permite tomar decisiones basadas en datos en áreas como:

* Control de calidad.
* Optimización de procesos.
* Desarrollo de productos.

🎯 **Ejemplo:**

> Una empresa desarrolla un sistema automático basado en visión computarizada para detectar defectos en empaques de productos alimenticios (como roturas, manchas, o etiquetas mal colocadas).
Para validar el sistema, se comparan los resultados (medidas de precisión de detección) en tres líneas de producción distintas que operan bajo condiciones ligeramente diferentes.

📊 **Diseño del experimento:**

* Variable dependiente (VD): Precisión (%) del sistema de visión en detectar correctamente los defectos.
* Variable independiente (VI): Línea de producción (A, B, C).
* Hipótesis nula (H₀): No hay diferencia significativa en la precisión del sistema entre las líneas.

#### **Aplicación:**

````python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Simular un dataset
np.random.seed(42)

# Tres líneas de producción con diferente nivel de defectos detectados por un sistema automático
data = {
    'Linea': ['Línea A']*10 + ['Línea B']*10 + ['Línea C']*10,
    'DefectosDetectados': np.concatenate([
        np.random.normal(5, 1, 10),   # Línea A
        np.random.normal(7, 1, 10),   # Línea B
        np.random.normal(6.5, 1, 10)  # Línea C
    ])
}

df = pd.DataFrame(data)

# Visualizar
plt.figure(figsize=(8, 5))
sns.boxplot(x='Linea', y='DefectosDetectados', data=df, hue='Linea', palette="Set2", legend=False)
plt.title('Defectos detectados por línea de producción')
plt.ylabel('Número de defectos detectados')
plt.show()

# ANOVA
modelo = ols('DefectosDetectados ~ C(Linea)', data=df).fit()
anova_tabla = sm.stats.anova_lm(modelo, typ=2)
print("\n📊 Tabla ANOVA:\n")
print(anova_tabla)

# Tukey HSD para comparar pares
tukey = pairwise_tukeyhsd(endog=df['DefectosDetectados'], groups=df['Linea'], alpha=0.05)
print("\n Resultados de Tukey HSD:\n")
print(tukey)

# Graficar resultados de Tukey
tukey.plot_simultaneous(comparison_name='Línea A')
plt.title('Comparaciones múltiples: Tukey HSD')
plt.show()
````

#### **Interpretación:**
* **ANOVA** te dice si existe alguna diferencia significativa entre las líneas de producción respecto a la cantidad de defectos detectados automáticamente.
* **Tukey HSD** revela entre qué líneas existen esas diferencias (por ejemplo, si la Línea A tiene menos defectos que la B o C).
* Esto es útil en validación de sistemas automáticos para verificar si el sistema detecta diferencias reales entre procesos productivos.

## 4.1. Ejemplo 1: Interpolación con Matriz de Vandermonde


### 📘 Fundamento Teórico

La **interpolación polinómica** busca construir una función que pase exactamente por un conjunto de puntos conocidos \((x_i, y_i)\). Una forma sistemática de hacerlo es utilizando la **Matriz de Vandermonde**, una matriz cuadrada en la que cada fila corresponde a una potencia del valor \(x_i\).

La matriz se define como:

<div align="center">
$
V = \begin{bmatrix}
1 & x_1 & x_1^2 & \cdots & x_1^{n-1} \\
1 & x_2 & x_2^2 & \cdots & x_2^{n-1} \\
\vdots & \vdots & \vdots & & \vdots \\
1 & x_n & x_n^2 & \cdots & x_n^{n-1}
\end{bmatrix}
$
</div>

Al resolver el sistema lineal

<div align="center">
$V \cdot a = y$
</div>

se obtienen los coeficientes del polinomio interpolante.

### 🎯 Aplicación Cuantitativa

Este método permite generar funciones que aproximan datos experimentales, útil en calibración, diseño de curvas, o ingeniería de control. Su implementación es completamente determinista y cuantitativa.


### 🔍 Implementación en Python

> Estimar un polinomio que pase por un conjunto de puntos conocidos utilizando álgebra lineal. Este método es fundamental para aproximaciones numéricas, modelado físico, y reconstrucción de señales.

````python
import numpy as np
import matplotlib.pyplot as plt

# Nuevos puntos
x = np.array([1, 2.5, 4])
y = np.array([3, 6, 10])
V = np.vander(x, increasing=True)
coefs = np.linalg.solve(V, y)

print("Coeficientes del polinomio interpolado:", coefs)

x_plot = np.linspace(0.5, 4.5, 100)
y_plot = coefs[0] + coefs[1]*x_plot + coefs[2]*x_plot**2

plt.plot(x_plot, y_plot, label='Polinomio interpolado')
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.title("Interpolación (versión de evaluación)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

### 📊 **Interpretación**

> Este ejercicio demuestra cómo el análisis cuantitativo permite construir un modelo determinístico (el polinomio) a partir de datos discretos. Es especialmente útil en simulaciones científicas donde se requieren aproximaciones suaves de datos experimentales.

## 4.2. Ejemplo 2: Ajuste de Regresión Lineal Simple

### 📘 Fundamento Teórico

La **regresión lineal** es una técnica estadística que modela la relación entre una variable dependiente $( y )$ y una o más variables independientes $( x )$ asumiendo que la relación es lineal:

<div align="center">
$y = \beta_0 + \beta_1 x + \epsilon$
</div>

donde:

- $( \beta_0 )$ es la ordenada al origen (intercepto),
- $( \beta_1 )$ es la pendiente (coeficiente que mide el cambio esperado en $( y )$ por unidad de cambio en $( x )$),
- $( \epsilon )$ es el término de error aleatorio.

El objetivo es encontrar los coeficientes $( \beta_0 )$ y $( \beta_1 )$ que minimizan la suma de los cuadrados de los residuos (diferencias entre valores observados y valores predichos).

### 🎯 Aplicación Cuantitativa

La regresión lineal se usa para:

- Predecir valores futuros o desconocidos,
- Evaluar la fuerza y dirección de la relación entre variables,
- Identificar tendencias y patrones en datos experimentales o económicos.

Este método es fundamental en análisis cuantitativo por su simplicidad y capacidad para inferir relaciones causales.


### 🔍 Implementación en Python

Estimar la relación lineal entre dos variables numéricas mediante un modelo de regresión lineal simple. Este método es fundamental en estadística, ciencia de datos e ingeniería para modelar tendencias, predecir valores y verificar hipótesis causales.

> Modelo general:

<div align="center">
$y = \beta_0 + \beta_1 x + \varepsilon$
</div>

donde:

- $y$ es la variable dependiente (respuesta),
- $x$ es la variable independiente (predictora),
- $\beta_0$ es la ordenada al origen (intercepto),
- $\beta_1$ es la pendiente,
- $\varepsilon$ es el error aleatorio.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm

# Simulación de datos con relación lineal
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 3 * x + 4 + np.random.normal(0, 2, size=len(x))

# Crear DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Ajustar modelo de regresión lineal
X = sm.add_constant(df['x'])  # Agrega el término constante (intercepto)
modelo = sm.OLS(df['y'], X).fit()

# Mostrar resumen del modelo
print(modelo.summary())

# Visualización
plt.figure(figsize=(8, 5))
sns.regplot(x='x', y='y', data=df, line_kws={"color": "red"}, ci=95)
plt.title("Regresión Lineal: Relación entre x e y")
plt.xlabel("Variable Independiente (x)")
plt.ylabel("Variable Dependiente (y)")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 📊 Interpretación

* El modelo estima los coeficientes $\beta_0$ y $\beta_1$ que mejor ajustan una línea recta a los datos.
* La salida de $summary()$ muestra:
    - Coeficientes estimados, error estándar, estadístico t y p-valor.
    - $R^2$, que indica qué proporción de la varianza de $y$ es explicada por $x$.
    - $p-value$ bajo $(< 0.05)$ indica que la relación entre las variables es estadísticamente significativa.
    - La gráfica muestra el ajuste lineal y el intervalo de confianza.

## 4.3. Ejemplo 3: Solución de una Ecuación Diferencial Ordinaria (EDO) con el Método de Euler

### 📘 Fundamento Teórico

El **método de Euler** es uno de los métodos numéricos más simples para resolver ecuaciones diferenciales ordinarias (EDO) de la forma:

<div align="center">
$\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0$
</div>

Se aproxima la solución utilizando la fórmula recursiva:

<div align="center">
$y_{n+1} = y_n + h \cdot f(t_n, y_n)$
</div>

donde \(h\) es el tamaño del paso.

### 🎯 Aplicación Cuantitativa

El método de Euler proporciona una aproximación paso a paso de la solución de un modelo dinámico. Su simplicidad lo hace útil para visualizar rápidamente la evolución de sistemas, con validación gráfica y numérica.


### 🔍 Implementación en Python

> Resolver numéricamente una EDO de primer orden usando el método de Euler, una técnica fundamental en análisis numérico para modelar sistemas dinámicos, reacciones químicas, crecimiento, y más.

>EDO dada:
<div align="center">
$\frac{dy}{dt} = -3y, \quad y(0) = 2$
</div>

````python
import numpy as np
import matplotlib.pyplot as plt

# EDO: dy/dt = -3y, y(0) = 2
t0, y0 = 0, 2
h = 0.1
n = 25

t_vals = [t0]
y_vals = [y0]

for _ in range(n):
    y_next = y_vals[-1] - 3 * y_vals[-1] * h
    t_next = t_vals[-1] + h
    y_vals.append(y_next)
    t_vals.append(t_next)

# Solución exacta
t_exact = np.linspace(0, 2.5, 100)
y_exact = 2 * np.exp(-3 * t_exact)

# Gráfico
plt.plot(t_vals, y_vals, 'o-', label='Euler (Aprox.)')
plt.plot(t_exact, y_exact, '--', label='Solución Exacta')
plt.title("EDO: dy/dt = -3y con Método de Euler")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
````

### 📊 Interpretación

* El método de Euler aproxima la solución de una EDO paso a paso, usando derivadas evaluadas en puntos discretos.
* Se observa que, para pasos pequeños (h = 0.1), la solución aproximada sigue de cerca la curva exacta.
* Esto ilustra cómo el análisis cuantitativo permite simular la evolución de sistemas dinámicos incluso sin una solución analítica directa.

## 4.4. Ejemplo 4: Modelo Logístico de Crecimiento Poblacional

### 📘 Fundamento Teórico

El modelo logístico describe el crecimiento de una población limitada por un entorno finito. Se basa en la ecuación diferencial:

<div align="center">
$\frac{dP}{dt} = rP \left(1 - \frac{P}{K}\right)$
</div>

donde:

- $(P)$ es la población en el tiempo $(t)$,
- $(r)$ es la tasa de crecimiento,
- $(K)$ es la capacidad máxima del entorno (capacidad de carga).

Este modelo refleja cómo la tasa de crecimiento disminuye a medida que la población se acerca a su límite.

### 🎯 Aplicación Cuantitativa

Es ampliamente utilizado en biología, ecología, economía y sociología para simular fenómenos de crecimiento limitado. Permite validar hipótesis sobre sostenibilidad y predicción de puntos de saturación.

### 🔍 Implementación en Python

> Simular la evolución de una población limitada por recursos, usando el modelo logístico, ampliamente utilizado en ecología, epidemiología, economía y sociología.

> Modelo:
<div align="center">
$\frac{dP}{dt} = rP\left(1 - \frac{P}{K}\right)$
</div>

````python
import matplotlib.pyplot as plt

# Parámetros
r = 0.3         # Tasa de crecimiento
K = 800         # Capacidad de carga
P0 = 50         # Población inicial
h = 0.2         # Paso de tiempo
n = 60          # Iteraciones

P_vals = [P0]
t_vals = [0]

for _ in range(n):
    P = P_vals[-1]
    dP = r * P * (1 - P / K)
    P_vals.append(P + dP * h)
    t_vals.append(t_vals[-1] + h)

plt.plot(t_vals, P_vals, label='Crecimiento logístico')
plt.title("Simulación de Crecimiento Poblacional (Modelo Logístico)")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.grid(True)
plt.show()
````

### 📊 Interpretación

* El modelo muestra cómo una población crece rápidamente al inicio y luego se estabiliza conforme se acerca al límite ambiental (K).
* Es un ejemplo clásico de dinámica no lineal y sirve para enseñar cómo los modelos cuantitativos pueden capturar comportamientos complejos con pocas variables y reglas simples.

## 4.5. Ejemplo 5: Análisis de Componentes Principales (PCA)

### 📘 Fundamento Teórico

El **Análisis de Componentes Principales (PCA)** es una técnica de reducción de dimensionalidad que transforma un conjunto de variables correlacionadas en un nuevo conjunto de variables no correlacionadas llamadas **componentes principales**.

Estas componentes están ordenadas según la varianza explicada: el primer componente explica la mayor parte de la varianza, el segundo la siguiente, y así sucesivamente.

### 🎯 Aplicación Cuantitativa

PCA permite analizar grandes conjuntos de datos de forma más eficiente, visualizar patrones ocultos, y eliminar redundancia. Es muy útil en ciencia de datos, ingeniería, y reconocimiento de patrones.

### 🔍 Implementación en Python

Reducir la dimensionalidad de un conjunto de datos multivariado para facilitar su análisis visual y estadístico, preservando la mayor parte de la varianza original.

> PCA transforma las variables originales en componentes principales no correlacionadas, ordenadas según su capacidad de explicar la varianza.

````python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos sintéticos
np.random.seed(123)
X = np.random.normal(loc=5, scale=2, size=(120, 4))
df = pd.DataFrame(X, columns=['A', 'B', 'C', 'D'])

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Proporción de varianza explicada:", pca.explained_variance_ratio_)

# Gráfico
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7, c='orange')
plt.title("PCA: Reducción de Dimensión (Versión Alternativa)")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid(True)
plt.show()
````

### 📊 Interpretación

* PCA identifica los ejes de mayor variabilidad en los datos, lo cual permite reducir dimensiones sin perder demasiada información.
* En este caso, las dos primeras componentes explican una parte sustancial de la varianza total, permitiendo visualizar en 2D datos originalmente en 4D.
* Es una técnica fundamental en el análisis cuantitativo moderno, útil en bioinformática, ingeniería, finanzas, IA y más.

plt.scatter(x, y, color='red', label='Puntos originales')
plt.xlabel('Tiempo (s)')
plt.ylabel('Temperatura (°C)')
plt.title('Interpolación con la Matriz de Vandermonde')
plt.legend()
plt.show()

\begin{lstlisting}[language=Python, caption=Modelo Logístico de Crecimiento Poblacional]
import numpy as np
import matplotlib.pyplot as plt

# Definición del modelo logístico
def logistic_model(t, P, r, K):
    return r * P * (1 - P / K)

# Parámetros
r = 0.1  # Tasa de crecimiento
K = 1000  # Capacidad de carga
P0 = 10  # Población inicial
t = np.linspace(0, 50, 500)

# Solución utilizando el modelo logístico
P = np.zeros(len(t))
P[0] = P0
for i in range(1, len(t)):
    P[i] = P[i-1] + logistic_model(t[i-1], P[i-1], r, K) * (t[i] - t[i-1])

# Gráfico del modelo logístico
plt.plot(t, P, label='Población')
plt.xlabel('Tiempo (t)')
plt.ylabel('Población P(t)')
plt.title('Crecimiento Poblacional - Modelo Logístico')
plt.legend()
plt.show()
\end{lstlisting}

## 6.1. Teorema 2.3 (Completo, de la página 32 de la tesis)

Sean $0 < h \leq \tau$ y $(m-1)\tau \leq t \leq m\tau - h$ para $m \geq 1$. Entonces, escribiendo $F(s) = \gamma \phi'(s) + \beta \phi(s)$, la solución $x(t)$ de (2.5), con condición inicial (2.2), satisface
$$x(t + h) = e^{\alpha h} \, x(t) + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{k-r} (\beta + \alpha \gamma)^r \frac{h^r}{r!} \binom{k-1}{r-1} x(t - k\tau)$$
$$+ \int_{t - m\tau}^{t - m\tau + h} \left[ \gamma^{m-1} + \sum_{k=1}^{m-1} \sum_{r=1}^{k} \gamma^{m-1-r} (\beta + \alpha \gamma)^r \frac{(t - m\tau + h - s)^r}{r!} \right] e^{\alpha (t - m\tau + h - s)} F(s) \, ds$$

El **Teorema 2.3** establece una relación estructural entre:

* Teoría científica
* Simulaciones computacionales
* Datos experimentales

Sirve como **eje epistemológico y cuantitativo** para validar experimentalmente en ciencia e ingeniería. Este marco permite justificar rigurosamente:

* Modelos predictivos (IA, regresión, PCA)
* Evaluación de hipótesis (ANOVA, Tukey)
* Simulaciones de sistemas complejos (EDO con retardo, qubits)
* Clasificación de datos y análisis exploratorio (Sturges, matriz de impacto)

## 6.3. Clasificación de Malware con Integración de Métodos Cuantitativos, Incluyendo EDR

El contexto: En ciberseguridad, el malware se propaga con retardos (e.j., tiempo de incubación en redes), lo que justifica modelar su crecimiento con EDR. Usaremos:

## 📁 **Paso 3: Importar librerías necesarias**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
````

## **⚙️ Paso 4: Leer el Dataset**

````python
df = pd.read_csv('Malware dataset.csv')  # Renómbralo si es necesario
````

## **🤖 Paso 5: Eliminar valor Nan**

````python
# Preprocesamiento: Eliminar columnas no numéricas/irrelevantes, manejar NaNs
df = df.drop(['hash'], axis=1)  # Hash no útil
le = LabelEncoder()
df['classification'] = le.fit_transform(df['classification'])  # 0: benign, 1: malware
df = df.fillna(0)  # Simple imputación
````

## **📊 Paso 6: Aplicación Regla de Sturges**

````python
# 1. Regla de Sturges: Binning para histograma de 'utime' (tiempo usuario)
n = len(df['utime'])
k = int(1 + 3.322 * np.log10(n))  # Número de bins por Sturges
plt.figure(figsize=(8, 6))
plt.hist(df['utime'], bins=k, edgecolor='black', color='skyblue')
plt.title('Histograma de utime con Regla de Sturges')
plt.xlabel('utime')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
print(f'Número de bins por Sturges: {k}')
print("Interpretación: La regla de Sturges optimiza la clasificación de datos en bins para revelar distribuciones, aquí mostrando una concentración en valores bajos de utime con posibles outliers en tiempos altos.")
print("Esto ayuda a entender la variabilidad en el uso de CPU por procesos, facilitando la detección de anomalías en malware.")
````

## **📉 Paso 7: Aplicación de PCA**

````python
# 2. PCA: Reducción a 5 componentes
features = df.drop('classification', axis=1)
pca = PCA(n_components=5)
pca_components = pca.fit_transform(features)
df_pca = pd.DataFrame(pca_components, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
df_pca['classification'] = df['classification']
plt.figure(figsize=(8, 6))
plt.bar(range(1, 6), pca.explained_variance_ratio_, color='lightgreen')
plt.title('Varianza Explicada por Componentes Principales (PCA)')
plt.xlabel('Componente Principal')
plt.ylabel('Proporción de Varianza')
plt.grid(True)
plt.show()
print(f'Varianza explicada: {pca.explained_variance_ratio_}')
print("Interpretación: PCA reduce la dimensionalidad capturando la varianza principal, aquí los primeros componentes explican la mayoría, simplificando el análisis de features complejas en el dataset.")
print("Esto revela correlaciones latentes entre métricas como memoria y tiempo, útiles para clasificar malware sin redundancias.")
````

## **🔥 Paso 8: Aplicación ANOVA + Tukey HSD**

````python
# 3. ANOVA + Tukey HSD: Comparar 'utime' entre clases (ANOVA para diferencias)
groups = [df['utime'][df['classification'] == i] for i in range(2)]
f_stat, p_val = stats.f_oneway(*groups)
plt.figure(figsize=(8, 6))
sns.boxplot(x='classification', y='utime', hue='classification', data=df, palette='Set3', legend=False)
plt.title('Boxplot de utime por Clase (Benigno vs. Malware)')
plt.xlabel('Clasificación (0: Benigno, 1: Malware)')
plt.ylabel('utime')
plt.grid(True)
plt.show()
print(f'ANOVA: F={f_stat}, p={p_val}')
# Tukey (para pares, aunque binario; extensible si más clases)
tukey = pairwise_tukeyhsd(df['utime'], df['classification'])
print(tukey)
print("Interpretación: ANOVA detecta diferencias significativas en medias de utime entre clases, con p-valor bajo indicando que malware tiende a valores distintos.")
print("Tukey confirma pares específicos, ayudando a validar si features discriminan efectivamente entre benigno y malware en análisis cuantitativo.")
````

## **🧠 Paso 9: Aplicación de Regresión Lineal**

````python
# 4. Regresión Lineal: Predecir 'utime' basado en 'stime' y 'gtime'
X = df[['stime', 'gtime']]
y = df['utime']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LinearRegression()
model.fit(X_train, y_train)
preds = model.predict(X_test)
mse = mean_squared_error(y_test, preds)
plt.figure(figsize=(8, 6))
plt.scatter(y_test, preds, color='orange', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title('Predicciones vs. Valores Reales en Regresión Lineal')
plt.xlabel('utime Real')
plt.ylabel('utime Predicho')
plt.grid(True)
plt.show()
print(f'Regresión Lineal MSE: {mse}')
print("Interpretación: La regresión lineal modela relaciones lineales entre variables, aquí prediciendo utime con bajo MSE, indicando buena ajuste.")
print("El gráfico muestra alineación cercana a la línea ideal, revelando dependencias en tiempos de sistema para predecir comportamientos de malware.")
````

## **🧪 Paso 10: Matriz de Impacto**

````python
# 5. Matriz de Probabilidad de Impacto: Clasificar riesgos (simple 3x3)
# Probabilidad: clasificación (0-1), Impacto: normalizar 'total_vm' (memoria)
df['prob'] = df['classification']  # 0 bajo, 1 alto
df['impact'] = pd.cut(df['total_vm'], bins=3, labels=[1,2,3])  # Bajo=1, Medio=2, Alto=3
def risk_level(prob, impact):
    if prob == 0: return 'Bajo'
    elif impact <= 1: return 'Bajo'
    elif impact == 2: return 'Medio'
    else: return 'Alto'
df['risk'] = df.apply(lambda row: risk_level(row['prob'], row['impact']), axis=1)
risk_counts = df['risk'].value_counts()
plt.figure(figsize=(8, 6))
risk_counts.plot(kind='bar', color=['green', 'yellow', 'red'])
plt.title('Distribución de Niveles de Riesgo (Matriz de Probabilidad de Impacto)')
plt.xlabel('Nivel de Riesgo')
plt.ylabel('Número de Muestras')
plt.grid(True)
plt.show()
print(df['risk'].value_counts())  # Matriz resumida
print("Interpretación: La matriz clasifica riesgos combinando probabilidad e impacto, aquí mostrando predominio de riesgos bajos pero identificando altos en malware.")
print("Esto cuantifica amenazas potenciales, guiando priorización en ciberseguridad con visualización clara de distribuciones.")
````

## **🧪 Paso 11: Modelo logístico simple**

````python
# 6. EDO con Euler: Modelo logístico simple I'(t) = r I (1 - I/K)
def euler_logistic(r=0.1, K=1000, I0=1, t_max=100, h=0.1):
    t = np.arange(0, t_max, h)
    I = np.zeros(len(t))
    I[0] = I0
    for i in range(1, len(t)):
        I[i] = I[i-1] + h * r * I[i-1] * (1 - I[i-1]/K)
    return t, I
t, I = euler_logistic()
plt.figure(figsize=(8, 6))
plt.plot(t, I, color='blue')
plt.title('EDO Logístico con Euler (Crecimiento de Infecciones)')
plt.xlabel('Tiempo')
plt.ylabel('Número de Infecciones')
plt.grid(True)
plt.show()
print("Interpretación: El método de Euler resuelve numéricamente EDO, modelando crecimiento logístico que satura en K, simulando propagación de malware sin retardos.")
print("El gráfico ilustra la curva sigmoide típica, útil para predecir picos de infecciones en escenarios ideales.")
````

## **📌 Paso 12: Modelo propagación de malware**

````python
# 7. EDR con Euler Adaptado (inspirado en Teorema 2.3): I'(t) = r I(t) (1 - I(t)/K) - δ I(t - τ)
# Usamos aproximación discreta (τ = 5, δ=0.05, N=τ/h=50 si h=0.1)
def euler_dde(r=0.1, K=1000, δ=0.05, τ=5, I0=1, t_max=100, h=0.1):
    t = np.arange(0, t_max, h)
    I = np.zeros(len(t))
    delay_steps = int(τ / h)
    for i in range(delay_steps):  # Inicializar con I0
        I[i] = I0
    for i in range(delay_steps, len(t)):
        delay_term = δ * I[i - delay_steps]  # Término retardado
        I[i] = I[i-1] + h * (r * I[i-1] * (1 - I[i-1]/K) - delay_term)
    return t, I
t_dde, I_dde = euler_dde()
plt.figure(figsize=(8, 6))
plt.plot(t_dde, I_dde, color='purple')
plt.title('EDR Logístico con Retardo (Modelando Propagación de Malware con Demora en Detección)')
plt.xlabel('Tiempo')
plt.ylabel('Número de Infecciones')
plt.grid(True)
plt.show()
print("Interpretación: Adaptando Euler a EDR (alineado al Teorema 2.3), incorpora retardos para modelar demoras reales en detección, resultando en oscilaciones o retrasos en el crecimiento.")
print("El gráfico muestra dinámica más realista con fluctuaciones, destacando impactos de retardos en predicciones de brotes de malware.")
````
