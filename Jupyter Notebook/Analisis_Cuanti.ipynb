# 1. Introducción

La epistemología es la rama de la filosofía que estudia el conocimiento: cómo se obtiene, valida y estructura.

 * **Definición:** La epistemología es el estudio de los fundamentos, la naturaleza y el alcance del conocimiento. En la investigación cuantitativa, se enfoca en cómo se construye el conocimiento a través de datos numéricos y cómo estos datos son procesados, analizados y validados.

* **Fundamento:** La investigación cuantitativa se basa en el paradigma positivista, que sostiene que la realidad es objetiva y puede ser medida de manera precisa a través de variables observables y manipulables.

## 1.2. Epistemología de la Investigación cuantitativa

🧠 **En el enfoque cuantitativo, el conocimiento se construye a partir de:**

* Observación empírica de fenómenos medibles.
* Uso de variables cuantificables.
* Aplicación de modelos matemáticos y estadísticos.

🔬 **Paradigma Positivista**

Supone que la realidad existe de forma objetiva e independiente del sujeto que la estudia.

* El conocimiento se obtiene por medio de la observación sistemática y el análisis lógico.
* Las hipótesis deben comprobarse con datos y los resultados deben ser replicables y generalizables.

<div align="center">
    
</div>

✅ **Técnicas como ANOVA, Tukey y PCA responden a este paradigma porque permiten:**

* Medir fenómenos con precisión.
* Identificar patrones y relaciones.
* Confirmar hipótesis con rigor estadístico.

<a id="1.2.1"></a>
### 1.2.1. Confiabilidad y validez

#### 🔁 **Confiabilidad**

Es la consistencia de los resultados al repetir un estudio o aplicar el mismo instrumento varias veces.

* Prueba-reprueba: Evalúa si los resultados se mantienen estables a lo largo del tiempo.
* Consistencia interna: Evalúa si los ítems dentro de un instrumento miden el mismo constructo.

#### 🎯 **Validez**

Es la capacidad de un instrumento o estudio para medir lo que realmente pretende medir.

* **_Validez interna:_** Capacidad del diseño para atribuir los cambios observados a la variable independiente.
* **_Validez externa:_** Grado en que los resultados se pueden generalizar a otras poblaciones o contextos.
* **_Validez de constructo:_** Relación entre los conceptos teóricos y las mediciones empíricas.

> <div align="center">

💡 **Por ejemplo, en un ANOVA bien diseñado:**

* La _confiabilidad_ se garantiza con réplicas y control experimental.
* La _validez interna_ se alcanza al controlar las variables extrañas.
* La _validez de constructo_ se refuerza con el diseño adecuado de las mediciones.

<a id="1.2.2"></a>
### 1.2.2. Identificación de variables

Toda investigación cuantitativa requiere:

* **_Variable independiente (VI):_** Es la que el investigador manipula (por ejemplo: tipo de tratamiento).
* **_Variable dependiente (VD):_** Es el resultado observado o medido (por ejemplo: crecimiento bacteriano).
* **_Variables controladas:_** Factores que se mantienen constantes para evitar sesgo.

<div align="center">


✔️ En los análisis como **PCA**, las variables cuantitativas pueden ser múltiples (pH, temperatura, densidad, color), y se busca entender cómo se relacionan entre sí.

✔️ En **ANOVA**, la VI podría ser el grupo de tratamiento y la absorbancia la medición obtenida.

<a id="1.2.3"></a>
### 1.2.3. Alcances de la Investigación Cuantitativa

* **_Generalización:_** Se buscan conclusiones extrapolables a una población mayor.
* **_Causalidad:_** Especialmente en diseños experimentales, permite establecer relaciones causa-efecto.
* **_Replicabilidad:_** Los procedimientos pueden repetirse en otros estudios.

🎓 **Las técnicas estadísticas como PCA, ANOVA y Tukey:**

* Refuerzan la replicabilidad, al seguir métodos estándar.
* Permiten interpretar fenómenos complejos mediante modelos matemáticos.
* Aportan evidencia empírica para validar teorías o hipótesis en ciencia e ingeniería.

# 2. Tipos de investigación cuantitativa

## 2.1. Investigación Experimental

Es el tipo más riguroso de investigación cuantitativa. Se caracteriza por:

* Manipulación directa de la VI.
* Control estricto de las variables externas.
* Asignación aleatoria de sujetos a grupos.

🎯 Su objetivo es establecer relaciones de **causalidad**.

🔬 **Ejemplo:**

> Evaluar el efecto de un conservante natural (como extracto vegetal) en la inhibición del crecimiento bacteriano, midiendo la absorbancia (VD) a lo largo de los días.

✅ El uso de **ANOVA** permite saber si existen diferencias significativas entre los grupos. Luego, Tukey HSD revela entre qué grupos están esas diferencias.

## 2.2. Investigación Cuasi-experimental

Se utiliza cuando no es posible aplicar aleatoriedad completa. Por ejemplo:

* Grupos ya formados (escuelas, plantas industriales, comunidades).
* Menor control de variables externas, pero aún se manipula la VI.


👨‍🏫 **Ejemplo:**

> Comparar el rendimiento de productos procesados en distintas líneas de producción, sin posibilidad de asignar condiciones aleatoriamente.

🔎 Se puede aplicar **ANOVA de medidas repetidas** o modelos mixtos para controlar parcialmente los factores no aleatorizados.

## 2.3. Investigación Cuantitativa No Experimental

No hay manipulación de variables. Se basa en la observación, registro y análisis estadístico de fenómenos existentes.

* **_Descriptiva:_** Describe características de una población o fenómeno.
* **_Correlacional:_** Examina relaciones entre variables sin establecer causalidad.

💧 **Ejemplo:**

> Analizar la relación entre turbidez, conductividad y pH en cuerpos de agua de distintas regiones.

🧠 Aquí, **PCA** es útil para reducir variables y encontrar las más relevantes o representativas del comportamiento global.

<div align="center">


🧩 **iNTERPRETACIÓN:**

* PC1 y PC2 son combinaciones de las variables originales (turbidez, conductividad, pH).
* Los grupos separados por región indican que ciertas regiones tienen características fisicoquímicas distintas.
* Puedes identificar qué variables contribuyen más a cada componente analizando pca.components_.

# 3. Ejemplos Prácticos

<a id="3.1"></a>
## 3.1. Ejemplo 1: PCA (Análisis de Componentes Principales)

* Se aplica cuando hay muchas variables y se desea encontrar patrones.
* Permite reducir la dimensionalidad sin perder información relevante.

Se usa para:
* Evaluar calidad de productos.
* Detectar variables redundantes.
* Visualizar agrupaciones naturales.

🧪 **Ejemplo aplicado:**

> Evaluar las propiedades fisicoquímicas de bebidas fermentadas con frutas andinas. PCA mostró cómo se agrupan según su perfil químico.

<div align="center">


<a id="3.2"></a>
## 3.2. Ejemplo 2: ANOVA + Tukey HSD

* **ANOVA** permite comparar medias de múltiples grupos.
* **Tukey HSD** identifica entre cuáles grupos hay diferencias significativas.

👨‍🔬 **Ejemplo aplicado:**

> Estudio de concentración de un compuesto natural en muestras de alimentos durante distintos días de almacenamiento. Se aplicó ANOVA por día y grupo de tratamiento; luego, Tukey mostró qué tratamientos conservaban mejor las propiedades.

<a id="3.3"></a>
## 3.3. Ejemplo 3: Aplicación en Ciencia e Ingeniería

* Uso de estas técnicas para validar sensores, materiales, procesos industriales.

Permite tomar decisiones basadas en datos en áreas como:

* Control de calidad.
* Optimización de procesos.
* Desarrollo de productos.

🎯 **Ejemplo:**

> Una empresa desarrolla un sistema automático basado en visión computarizada para detectar defectos en empaques de productos alimenticios (como roturas, manchas, o etiquetas mal colocadas).
Para validar el sistema, se comparan los resultados (medidas de precisión de detección) en tres líneas de producción distintas que operan bajo condiciones ligeramente diferentes.

📊 **Diseño del experimento:**

* Variable dependiente (VD): Precisión (%) del sistema de visión en detectar correctamente los defectos.
* Variable independiente (VI): Línea de producción (A, B, C).
* Hipótesis nula (H₀): No hay diferencia significativa en la precisión del sistema entre las líneas.

#### **Aplicación:**

````python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Simular un dataset
np.random.seed(42)

# Tres líneas de producción con diferente nivel de defectos detectados por un sistema automático
data = {
    'Linea': ['Línea A']*10 + ['Línea B']*10 + ['Línea C']*10,
    'DefectosDetectados': np.concatenate([
        np.random.normal(5, 1, 10),   # Línea A
        np.random.normal(7, 1, 10),   # Línea B
        np.random.normal(6.5, 1, 10)  # Línea C
    ])
}

df = pd.DataFrame(data)

# Visualizar
plt.figure(figsize=(8, 5))
sns.boxplot(x='Linea', y='DefectosDetectados', data=df, hue='Linea', palette="Set2", legend=False)
plt.title('Defectos detectados por línea de producción')
plt.ylabel('Número de defectos detectados')
plt.show()

# ANOVA
modelo = ols('DefectosDetectados ~ C(Linea)', data=df).fit()
anova_tabla = sm.stats.anova_lm(modelo, typ=2)
print("\n📊 Tabla ANOVA:\n")
print(anova_tabla)

# Tukey HSD para comparar pares
tukey = pairwise_tukeyhsd(endog=df['DefectosDetectados'], groups=df['Linea'], alpha=0.05)
print("\n Resultados de Tukey HSD:\n")
print(tukey)

# Graficar resultados de Tukey
tukey.plot_simultaneous(comparison_name='Línea A')
plt.title('Comparaciones múltiples: Tukey HSD')
plt.show()
````

#### **Interpretación:**
* **ANOVA** te dice si existe alguna diferencia significativa entre las líneas de producción respecto a la cantidad de defectos detectados automáticamente.
* **Tukey HSD** revela entre qué líneas existen esas diferencias (por ejemplo, si la Línea A tiene menos defectos que la B o C).
* Esto es útil en validación de sistemas automáticos para verificar si el sistema detecta diferencias reales entre procesos productivos.
