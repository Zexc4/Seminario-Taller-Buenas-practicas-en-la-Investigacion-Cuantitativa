```python
# 📦 Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy import stats

# ✅ Simulación de datos para PCA
np.random.seed(0)
n = 100

# Cuatro variables cuantitativas (pH, temperatura, densidad, color)
pH = np.random.normal(7, 0.5, n)
temperatura = np.random.normal(25, 2, n)
densidad = np.random.normal(1.0, 0.05, n)
color = np.random.normal(50, 10, n)

# Crear DataFrame
df_pca = pd.DataFrame({
    'pH': pH,
    'Temperatura': temperatura,
    'Densidad': densidad,
    'Color': color
})

# 🔍 PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pca)
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)

# 📊 Gráfico PCA
plt.figure(figsize=(8, 6))
plt.scatter(components[:, 0], components[:, 1], c='steelblue', edgecolor='k')
plt.title('PCA - Relación entre variables cuantitativas')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```
